```
#Số amstrong
def isAmstrong(n, k):
    x = int(n)
    sum = 0
    for i in range(k):   
        sum += pow(x % 10, k)
        x = x//10
    if sum == int(n): return True
    return False

#main
n = input()
k = len(n)
print(isAmstrong(n, k))
```

```
#Xử lý chuỗi
def isVowel(c):
    if c == 'a': return True
    if c == 'e': return True
    if c == 'i': return True
    if c == 'o': return True
    if c == 'u': return True
    if c == 'y': return True
    return False

s = input().casefold()
txt = ''
for i in range(len(s)):
    if isVowel(s[i]) == False:
        txt += '.' + s[i]
print(txt)
```

```
#Ngôn ngữ của Lan
def IsLanLanguage(s, noun):
    #check adj
    for i in range(noun):
        tmp = s[i]
        if ("lios" != tmp[-4:]) and ("liala" != tmp[-5:]):
            return("NO")
    #check verb
    for i in range(noun + 1, len(s)):
        tmp = s[i]
        if ("initis" != tmp[-6:]) and ("inites" != tmp[-6:]):
            return("NO")
    return("YES")

s = input().split()
noun = 0
for i in s:
    #check noun
    if ("etr" in i) or ("etra" in i):
        break
    noun += 1 #find noun location
print(IsLanLanguage(s, noun))
```

```
#Rút gọn phân số
def GCD(a, b):
    if b == 0: return a
    return GCD(b, a % b)

def Simplify(a, b):
    tmp = GCD(a, b)
    print(a // tmp, b // tmp)
    if (b // tmp) > 1: print(a // tmp)

n = int(input())
for i in range(n):
    a, b = list([int(i) for i in input().split()])
    Simplify(a, b)
```

```
#Xếp que diêm
def MinMatchstick(n):
    if n == 2: return 2
    if n % 2 != 0: return 1
    return 0

q = int(input())
for i in range(q):
    n = int(input())
    print(MinMatchstick(n))
```

```
#Chuỗi bằng nhau
def IsEqual(s, t):
    for i in s:
        if i in t: return ("YES")
    for i in t:
        if i in s: return ("YES")
    return ("NO")

q = int(input())
for i in range(q):
    s = input()
    t = input()
    if s == t: print("YES")
    else: print(IsEqual(s, t))
```

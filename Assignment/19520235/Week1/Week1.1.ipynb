```
#Tam giác
from math import sqrt

def isTriangle(a, b, c):
    if (a + b <= c) or (b + c <= a) or (c + a <= b):
        return False
    return True

def Heron(a, b, c):
    p = (a + b + c)/2
    s = sqrt(p*(p - a)*(p - b)*(p - c))
    if (s - round(s, 2) == 0):
        return int(s)
    return round(s, 2)

def Classify(a, b, c):
    if (a == b) and (b == c): 
         return "Tam giac deu, dien tich =" 
    if (a == b) or (b == c) or (c == a):
        return "Tam giac can, dien tich ="
    if a*a == b*b + c*c or b*b == a*a + c*c or c*c== b*b + a*a:
        return "Tam giac vuong, dien tich ="
    return "Tam giac thuong, dien tich ="

#input
a = float(input())
b = float(input())
c = float(input())

#main
if isTriangle(a, b, c):
    print(Classify(a, b, c), Heron(a, b,c))
else:
    print("Khong phai tam giac")
```

```
#Task
def BobPos(n, k, p, q):
    AlicePos = 2*(p - 1) + q
    front = AlicePos - k #vị trí gần nhất trước Alice 
    back = AlicePos + k #vị trí gần nhất sau Alice 
    if front >= 1:
        return front
    if back <= n:
        return back
    return -1

#input
n = int(input())
k = int(input())
p = int(input())
q = int(input())

#main
rslt = BobPos(n, k, p, q)
if rslt == -1:
    print('-1')
elif rslt % 2 == 0:
    print(rslt // 2, '2')
else:
    print((rslt // 2) + 1, '1')
```
